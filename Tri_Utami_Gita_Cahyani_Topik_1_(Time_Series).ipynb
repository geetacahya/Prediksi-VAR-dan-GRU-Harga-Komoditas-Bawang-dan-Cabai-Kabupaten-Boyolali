### 2.2 Metode Analisis Data 1: Vector Autoregression (VAR)

Vector Autoregression (VAR) merupakan salah satu metode dalam statistika dan ekonomi kuantitatif yang digunakan untuk menganalisis hubungan dinamis antar beberapa variabel waktu (multivariate time series). VAR pertama kali diperkenalkan oleh Christopher A. Sims pada tahun 1980 sebagai alternatif dari model struktural simultan yang kompleks dalam ekonomi (Sims, 1980). Berbeda dengan regresi biasa yang membedakan variabel dependen dan independen secara tegas, VAR memperlakukan semua variabel sebagai endogen, sehingga memungkinkan setiap variabel untuk dipengaruhi oleh nilai masa lalu dari dirinya sendiri dan variabel lain dalam sistem.

Dalam model VAR, setiap variabel dalam sistem dimodelkan sebagai fungsi linear dari lag-nya sendiri dan lag dari semua variabel lain dalam sistem. Kelebihan utama dari VAR adalah kesederhanaan strukturnya dan fleksibilitasnya dalam menangkap dinamika interdependensi antara variabel ekonomi, sosial, atau komoditas (Gujarati & Porter, 2009).

Model VAR(*p*) (VAR orde *p*) untuk k variabel endogen dapat ditulis dalam bentuk matriks sebagai berikut:

\begin{equation}
\mathbf{Y}_t = \mathbf{c} + \mathbf{A}_1 \mathbf{Y}_{t-1} + \mathbf{A}_2 \mathbf{Y}_{t-2} + \dots + \mathbf{A}_p \mathbf{Y}_{t-p} + \boldsymbol{\varepsilon}_t
\end{equation}

dengan:
- $\mathbf{Y}_t$ adalah **vektor** $(k \times 1)$ dari variabel endogen pada waktu ke-$t$,
- $\mathbf{c}$ adalah **vektor konstanta** $(k \times 1)$,
- $\mathbf{A}_i$ adalah **matriks koefisien** $(k \times k)$ untuk lag ke-$i$, di mana $i = 1, 2, \dots, p$,
- $p$ adalah **orde lag** dari model VAR,
- $\boldsymbol{\varepsilon}_t$ adalah **vektor error** $(k \times 1)$ pada waktu ke-$t$ yang diasumsikan bersifat white noise, yaitu:
  
\begin{equation}
\boldsymbol{\varepsilon}_t \sim \text{WN}(0, \boldsymbol{\Sigma})
\end{equation}

dimana $\boldsymbol{\Sigma}$ adalah **matriks kovarian** dari error.


Model Vector Autoregression (VAR) dipilih dalam penelitian ini karena mampu menangkap hubungan dinamis antar variabel multivariat dalam deret waktu. Dalam konteks harga komoditas seperti bawang dan cabai yang saling memengaruhi satu sama lain, pendekatan VAR sangat relevan karena memungkinkan setiap variabel diprediksi tidak hanya berdasarkan nilainya sendiri di masa lalu, tetapi juga nilai variabel lain. Metode ini tidak memerlukan asumsi tentang hubungan sebab-akibat secara eksplisit, sehingga cocok diterapkan ketika semua variabel dianggap saling memengaruhi. Selain itu, VAR juga memberikan fleksibilitas dalam menganalisis respons impuls (IRF) dan dekomposisi varians kesalahan (FEVD), yang penting untuk memahami keterkaitan jangka pendek dan panjang antar harga komoditas. Oleh karena itu, VAR dipilih sebagai salah satu pendekatan utama dalam memodelkan dan memprediksi harga komoditas di Kabupaten Boyolali.

#### Data Splitting

```python
# Split data: 85% train, 15% test
split_point = int(len(df) * 0.85)
df_train = df.iloc[:split_point]
df_test = df.iloc[split_point:]
```

Kode tersebut digunakan untuk membagi dataset menjadi data latih (training) dan data uji (testing) dengan proporsi 85% untuk pelatihan dan 15% untuk pengujian.

#### Uji Stasioneritas

```python
import matplotlib.pyplot as plt

# Ubah kolom Date menjadi datetime dan set sebagai index
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_train.set_index('Date', inplace=True)

# Buat subplots: 4 baris, 1 kolom
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(14, 10), sharex=True)

# Daftar kolom komoditas
columns = df_train.columns

# Plot masing-masing variabel
for i, col in enumerate(columns):
    axes[i].plot(df_train.index, df_train[col], label=col, color='tab:blue')
    axes[i].set_title(f"Tren Harga {col} di Kabupaten Boyolali Sebelum Stasioner", fontsize=12)
    axes[i].set_ylabel("Harga (Rp)")
    axes[i].legend(loc='upper left')
    axes[i].grid(True)

axes[-1].set_xlabel("Tanggal")
plt.tight_layout()
plt.show()
```

Gambar tersebut menunjukkan visualisasi tren harga empat komoditasâ€”bawang merah, bawang putih, cabai merah, dan cabai rawitâ€”di Kabupaten Boyolali sebelum dilakukan transformasi untuk mencapai stasioneritas. Terlihat bahwa keempat komoditas memiliki pola fluktuasi dan tren yang berbeda-beda sepanjang waktu, yang menandakan bahwa data belum stasioner karena masih terdapat tren naik atau turun, serta perubahan variabilitas. Oleh karena itu, diperlukan transformasi seperti differencing agar data dapat digunakan dalam pemodelan VAR.

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series, name):
    result = adfuller(series)
    return {
        "Komoditas": name,
        "ADF Statistic": result[0],
        "p-value": result[1],
        "Stationary": result[1] < 0.05
    }

# Uji ADF
adf_results = [adf_test(df_train[col], col) for col in df_train.columns]
adf_df = pd.DataFrame(adf_results)
adf_df
```

Hasil uji Augmented Dickey-Fuller (ADF) menunjukkan bahwa seluruh seri waktu untuk komoditas bawang merah, bawang putih, cabai merah, dan cabai rawit memiliki nilai p-value di atas 0,05. Ini berarti data belum stasioner secara statistik pada level signifikansi 5%, sehingga perlu dilakukan transformasi seperti diferensiasi agar data menjadi stasioner sebelum dilakukan pemodelan lebih lanjut.

```python
# Differencing
df_diff = df_train.diff().dropna()

# Uji ADF setelah differencing
adf_diff_results = [adf_test(df_diff[col], col) for col in df_diff.columns]
adf_diff_df = pd.DataFrame(adf_diff_results)
adf_diff_df
```

Setelah dilakukan differencing, hasil uji ADF menunjukkan bahwa seluruh variabel komoditas memiliki p-value jauh di bawah 0,05, yang menandakan bahwa data telah menjadi stasioner. Hal ini memungkinkan data untuk digunakan dalam pemodelan deret waktu seperti VAR secara valid.

```python
# Buat subplots untuk data training yang sudah di-differencing
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(14, 10), sharex=True)

for i, col in enumerate(df_diff.columns):
    axes[i].plot(df_diff.index, df_diff[col], label=f'{col} (Differenced)', color='tab:orange')
    axes[i].set_title(f"Perubahan Harga {col} Setelah Differencing/Stasioner", fontsize=12)
    axes[i].set_ylabel("Harga (Rp)")
    axes[i].legend(loc='upper left')
    axes[i].grid(True)

axes[-1].set_xlabel("Tanggal")
plt.tight_layout()
plt.show()
```

Gambar tersebut menunjukkan hasil transformasi differencing terhadap data harga komoditas di Kabupaten Boyolali, yang dilakukan untuk menjadikan data bersifat stasioner. Setelah differencing, fluktuasi nilai menjadi lebih stabil di sekitar garis nol tanpa tren yang jelas, yang mengindikasikan bahwa data telah memenuhi asumsi stasioneritas dan siap digunakan untuk pemodelan VAR.

#### Penentuan Lag Optimal

Setelah data sudah stasioner, maka langkah selanjutnya adalah penentuan lag optimum untuk model VAR.

```python
from statsmodels.tsa.api import VAR

model = VAR(df_diff)
lag_order = model.select_order(maxlags=30)

print("Lag Order Selection:")
print(lag_order.summary())
```

```python
# Tampilkan Hasil
print("\nLag optimal masing-masing kriteria:")
print(f"AIC  : {lag_order.aic}")
print(f"BIC  : {lag_order.bic}")
print(f"FPE  : {lag_order.fpe}")
print(f"HQIC : {lag_order.hqic}")
```

Hasil seleksi lag menunjukkan bahwa kriteria AIC, FPE, dan HQIC merekomendasikan lag optimal sebanyak 1, sementara BIC merekomendasikan lag 0. Karena sebagian besar kriteria memilih lag 1, maka lag 1 dapat dipertimbangkan sebagai pilihan terbaik untuk pemodelan VAR pada data ini.

#### Estimasi Parameter Untuk Model VAR (Metode OLS)

Pada tahap ini akan dicari estimasi model VAR. Adapun metode yang digunakan pada penelitian ini adalah metode OLS atau metode kuadrat terkecil dengan mendapatkan model dengan residual minimum dilakukan dengan cara mengkuadratkan residual maka
diperoleh estimasi parameter Î². Hasil yang diperoleh untuk estimasi parameternya melalui OLS ditampilkan pada output dibawah ini.

```python
# Fit model dengan lag optimal
model_fitted = model.fit(lag_order.aic)
print(model_fitted.summary())
```

Model VAR dengan lag 1 berhasil diestimasi menggunakan metode OLS untuk empat variabel harga komoditas. Berdasarkan hasil estimasi, hanya beberapa koefisien yang signifikan secara statistik pada taraf 5%. Misalnya, harga cabai rawit pada lag sebelumnya memiliki pengaruh signifikan terhadap harga cabai merah (p-value = 0.004) dan terhadap dirinya sendiri (p-value = 0.000), menunjukkan adanya keterkaitan kuat dalam dinamika harga cabai rawit. Selain itu, harga cabai merah juga menunjukkan pengaruh marginal signifikan terhadap dirinya sendiri dengan p-value = 0.050. Sementara itu, variabel lainnya sebagian besar tidak signifikan, menunjukkan bahwa pengaruh harga komoditas lain terhadap perubahan harga komoditas tertentu cenderung lemah. Korelasi residual antar variabel relatif rendah, kecuali antara cabai merah dan cabai rawit yang menunjukkan korelasi cukup kuat sebesar 0.56. Hal ini mencerminkan kemungkinan keterkaitan pola pergerakan harga antara kedua komoditas tersebut.

#### Uji Kausalitas Granger

Pengujian Granger Causality digunakan untuk mengetahui hubungan antar variabel apakah
terdapat hubungan searah atau kausal (dua arah).Berikut ini merupakan hasil pengujian
Granger Causality antara variabel.

```python
from statsmodels.tsa.stattools import grangercausalitytests

lag = 1  # Sesuai hasil seleksi lag

print("Hasil Uji Kausalitas Granger antar semua pasangan variabel:\n")
for y in df_diff.columns:
    for x in df_diff.columns:
        if x != y:
            print(f"Apakah '{x}' Granger menyebabkan '{y}'?")
            grangercausalitytests(df_diff[[y, x]], maxlag=lag, verbose=True)
            print("\n" + "-"*70 + "\n")
```

Hasil uji kausalitas Granger menunjukkan bahwa sebagian besar pasangan variabel tidak memiliki hubungan kausal secara statistik pada lag 1, ditunjukkan oleh p-value yang lebih besar dari 0,05. Artinya, nilai masa lalu suatu komoditas umumnya tidak cukup kuat untuk memprediksi pergerakan harga komoditas lainnya. Namun, terdapat satu hubungan yang signifikan, yaitu harga **cabai rawit** secara Granger menyebabkan harga **cabai merah**, dengan p-value sekitar 0,003. Ini menunjukkan bahwa perubahan harga cabai rawit di masa lalu memiliki kontribusi yang signifikan dalam menjelaskan variasi harga cabai merah saat ini. Hubungan ini mencerminkan adanya keterkaitan yang erat antara kedua komoditas tersebut, kemungkinan karena kesamaan fungsi pasar atau pola konsumsi.


#### Impulse Response Function (IRF)

Pada Bawang Putih, proporsi terbesar dari kesalahan prediksi juga dijelaskan oleh dirinya sendiri, yakni sekitar 95,6% hingga 96% selama lima periode. #### In-sample Prediction (Fitted Values)

Langkah ini digunakan untuk mengembalikan hasil prediksi in-sample dari skala differencing ke skala asli (harga sebenarnya). Nilai prediksi differencing dijumlahkan secara kumulatif (cumsum) lalu ditambahkan ke nilai harga terakhir sebelum prediksi dimulai, yaitu last_known. Dengan begitu, diperoleh estimasi harga aktual harian berdasarkan hasil prediksi model VAR.

### 2.3 Metode Analisis Data 2: Gated Recurrent Unit (GRU)

Gated Recurrent Unit (GRU) merupakan salah satu varian dari Recurrent Neural Network (RNN) yang dirancang untuk mengatasi kelemahan RNN tradisional, khususnya terkait vanishing gradient problem. GRU diperkenalkan oleh Cho et al. (2014) sebagai alternatif yang lebih sederhana dibandingkan Long Short-Term Memory (LSTM), dengan jumlah parameter yang lebih sedikit namun tetap mampu menangkap dependensi jangka panjang dalam data deret waktu. GRU menggabungkan gating mechanism yang memungkinkan model untuk mempertahankan informasi penting dari waktu ke waktu serta melupakan informasi yang tidak relevan. Dua komponen utama dalam GRU adalah update gate dan reset gate, yang bekerja sama untuk mengontrol aliran informasi.

\begin{equation}
z_t = \sigma(W_z \cdot x_t + U_z \cdot h_{t-1})
\end{equation}
\begin{equation}
r_t = \sigma(W_r \cdot x_t + U_r \cdot h_{t-1})
\end{equation}
\begin{equation}
\tilde{h}_t = \tanh(W_h \cdot x_t + U_h \cdot (r_t \odot h_{t-1}))
\end{equation}
\begin{equation}
h_t = (1 - z_t) \odot h_{t-1} + z_t \odot \tilde{h}_t
\end{equation}

dengan:
- $x_t$ adalah input pada waktu ke-$t$
- $h_{t-1}$ adalah hidden state sebelumnya
- $h_t$ adalah hidden state saat ini
- $z_t$ adalah update gate
- $r_t$ adalah reset gate
- $\tilde{h}_t$ adalah kandidat hidden state
- $\odot$ adalah operasi perkalian elemen-per-elemen (Hadamard product)
- $\sigma$ adalah fungsi aktivasi sigmoid
- $\tanh$ adalah fungsi aktivasi tangens hiperbolik
- $W_z$, $W_r$, $W_h$ adalah bobot parameter dari input dan hidden state
- $U_z$, $U_r$, $U_h$ adalah bobot hidden state sebelumnya untuk masing-masing gate dan kandidat state

Metode Gated Recurrent Unit (GRU) dipilih karena kemampuannya dalam menangani data deret waktu yang memiliki ketergantungan temporal, seperti harga komoditas yang cenderung dipengaruhi oleh nilai-nilai sebelumnya. Berbeda dengan model regresi linear atau model statistik klasik lainnya, GRU mampu menangkap pola nonlinier dan hubungan jangka panjang antar waktu secara lebih efektif. Selain itu, GRU lebih efisien secara komputasional dibandingkan Long Short-Term Memory (LSTM), karena hanya menggunakan dua gerbang utamaâ€”update gate dan reset gateâ€”sehingga jumlah parameter yang perlu dilatih lebih sedikit. Hal ini menjadikan GRU sebagai pilihan yang tepat untuk memodelkan data berurutan dengan struktur kompleks namun tetap menjaga efisiensi waktu pelatihan. Dalam konteks prediksi harga komoditas seperti bawang dan cabai, GRU memungkinkan model untuk belajar dari dinamika data masa lalu secara adaptif tanpa memerlukan arsitektur jaringan yang terlalu rumit.

Berikut tahap-tahap analisis data dengan metode Gated Recurrent Unit (GRU) berdasarkan kode dan proses yang telah dilakukan:

1. Mengimpor data harga komoditas bawang dan cabai dalam format CSV dan menampilkan lima baris awal untuk memastikan struktur data telah terbaca dengan benar.

2. Melakukan normalisasi data menggunakan metode MinMaxScaler ke dalam rentang \[0, 1] untuk seluruh variabel target, yaitu Bawang Merah, Bawang Putih, Cabai Merah, dan Cabai Rawit, agar data berada pada skala yang sesuai untuk pelatihan model GRU.

3. Melakukan transformasi bentuk data menjadi format sekuensial dengan pendekatan jendela geser (sliding window) selama 20 hari untuk membentuk data input (X) dan target output (y) agar sesuai dengan kebutuhan input jaringan GRU multivariate.

4. Membagi data menjadi data pelatihan (train) sebanyak 85% dan data pengujian (test) sebanyak 15% dari total dataset yang telah ditransformasikan.

5. Membangun arsitektur model GRU multivariate menggunakan TensorFlow dengan menambahkan layer GRU sebanyak 200 unit, dilanjutkan dengan dropout sebesar 0.5 untuk mencegah overfitting, dan diakhiri dengan layer Dense sebanyak 4 neuron sebagai output.

6. Menyusun dan mengompilasi model GRU dengan fungsi loss Mean Squared Error dan optimizer Adam, serta menggunakan metrik evaluasi MAE dan MSE.

7. Melatih model GRU menggunakan data latih selama 20 epoch untuk mempelajari pola dari deret waktu multivariat yang mencakup keempat komoditas.

8. Melakukan prediksi terhadap data uji menggunakan model yang telah dilatih, lalu mentransformasikan hasil prediksi dan data aktual kembali ke skala asli menggunakan inverse transform.

9. Membuat visualisasi hasil prediksi dan data aktual untuk masing-masing komoditas (Bawang Merah, Bawang Putih, Cabai Merah, dan Cabai Rawit) dalam bentuk grafik garis berdasarkan indeks waktu.

10. Menghitung metrik evaluasi model berupa Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), dan Mean Absolute Percentage Error (MAPE) untuk masing-masing komoditas guna mengukur akurasi hasil prediksi terhadap data aktual.

#### Persiapan Data

```python
data = pd.read_csv('Harga Komoditas Bawang dan Cabai Kabupaten Boyolali.csv')
print("Harga Komoditas Bawang dan Cabai Kabupaten Boyolali")
data.head()
```

#### Pra-pemrosesan Data: Normalisasi untuk Model GRU

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
columns_to_scale = ['Bawang Merah', 'Bawang Putih', 'Cabai Merah', 'Cabai Rawit']
data_scaled_values = scaler.fit_transform(data[columns_to_scale].to_numpy())
data_scaled = pd.DataFrame(data_scaled_values, columns=columns_to_scale)

target_scaler = MinMaxScaler(feature_range=(0, 1))

data_scaled[['Bawang Merah', 'Bawang Putih',
             'Cabai Merah', 'Cabai Rawit']] = target_scaler.fit_transform (data[[
              'Bawang Merah', 'Bawang Putih', 'Cabai Merah', 'Cabai Rawit']].to_numpy())

data_scaled = data_scaled.astype(float)
print(data_scaled.head())
```

Kode tersebut melakukan proses normalisasi data terhadap empat kolom harga komoditas, yaitu Bawang Merah, Bawang Putih, Cabai Merah, dan Cabai Rawit menggunakan teknik Min-Max Scaling, agar semua fitur memiliki skala antara 0 hingga 1.

#### Transformasi Bentuk Data

```python
import numpy as np

def singleStepSampler (df, window):
  xRes = []
  yRes = []
  for i in range(0, len(df) - window):
    res = []
    for j in range(0, window):
      r = []
      for col in df.columns:
        r.append(df[col].iloc[i + j])
      res.append(r)
    xRes.append(res)
    yRes.append(df[['Bawang Merah', 'Bawang Putih', 'Cabai Merah', 'Cabai Rawit']].iloc[i + window].values)
  return np.array(xRes), np.array(yRes)

(xVal, yval) = singleStepSampler(data_scaled, 20)
```

Transformasi bentuk data dilakukan untuk mengubah data deret waktu menjadi format yang sesuai dengan input model GRU, yaitu berbentuk urutan data dengan panjang jendela waktu tertentu. Fungsi `singleStepSampler` membuat pasangan input-output, di mana setiap input berisi data selama 20 hari sebelumnya, dan output-nya adalah nilai pada hari ke-21. Hasil transformasi ini menghasilkan array `xVal` sebagai data input dan `yval` sebagai target prediksi.

#### Data Splitting

```python
import tensorflow as tf

# Dataset splitting
SPLIT = 0.85
X_train = xVal[:int(SPLIT * len(xVal))]
y_train = yval[:int (SPLIT * len(yval))]
X_test = xVal[int(SPLIT * len(xVal)):]
y_test = yval[int (SPLIT * len(yval)):]
multivariate_gru = tf.keras.Sequential()
multivariate_gru.add(tf.keras.layers.GRU (200, input_shape=(X_train.shape[1], X_train.shape[2])))
multivariate_gru.add(tf.keras.layers.Dropout(0.5))
```

Pembagian data dilakukan dengan memisahkan 85% data untuk pelatihan dan 15% sisanya untuk pengujian guna menguji kemampuan generalisasi model. Setelah pembagian, model GRU dibangun menggunakan arsitektur berurutan dengan satu lapisan GRU yang memiliki 200 unit untuk menangkap pola dalam urutan data, serta ditambahkan lapisan dropout sebesar 0.5 untuk mengurangi risiko overfitting selama pelatihan.

#### Pelatihan Model

```python
multivariate_gru.add(tf.keras.layers.Dense(4, activation='linear'))
multivariate_gru.compile(loss='MeanSquaredError', metrics=['MAE', 'MSE'],
                         optimizer=tf.keras.optimizers.Adam())
multivariate_gru.summary()
history = multivariate_gru.fit(X_train, y_train, epochs=20)
```

Model GRU dilengkapi dengan lapisan output Dense berisi 4 neuron untuk memprediksi masing-masing komoditas. Model dikompilasi menggunakan fungsi loss `MeanSquaredError`, metrik evaluasi `MAE` dan `MSE`, serta optimizer Adam yang umum digunakan untuk pelatihan jaringan saraf. Proses pelatihan dilakukan selama 20 epoch menggunakan data pelatihan, dengan total parameter yang dilatih sebanyak 124.404.

#### Prediksi

```python
# Prediksi
y_pred = multivariate_gru.predict(X_test)

# Invers transform ke skala asli (karena sebelumnya di-normalisasi)
y_test_inv = scaler.inverse_transform(y_test)
y_pred_inv = scaler.inverse_transform(y_pred)
```

```python
# Panjang data test
n_pred = y_pred.shape[0]
start_date = df.index[-n_pred]
```

Model yang telah dilatih digunakan untuk memprediksi data pengujian, kemudian hasil prediksi dan data aktual dikembalikan ke skala aslinya menggunakan inverse transform agar dapat dibandingkan secara nyata. Jumlah data yang diprediksi dihitung, dan indeks tanggal awal untuk visualisasi ditentukan berdasarkan panjang data pengujian.

#### Metrik Evaluasi Model VAR

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = pd.Series(y_true), pd.Series(y_pred)
    return (abs((y_true - y_pred) / y_true)
            .replace([float('inf'), -float('inf')], float('nan'))
            .dropna()).mean() * 100

for col in forecast_df_real.columns:
    y_true = df_test[col]
    y_pred = forecast_df_real[col]

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = mse ** 0.5
    mape = mean_absolute_percentage_error(y_true, y_pred)

    print(f'{col}')
    print(f'   MAE  : {mae:.2f}')
    print(f'   MSE  : {mse:.2f}')
    print(f'   RMSE : {rmse:.2f}')
    print(f'   MAPE : {mape:.2f}%\n')
```

Sementara itu, pada cabai merah dan cabai rawit, nilai MAE dan RMSE lebih tinggi dibandingkan bawang, dan nilai MAPE juga meningkat masing-masing menjadi sekitar 6.70% dan 5,37%. Meskipun begitu, tingkat kesalahan ini masih tergolong wajar mengingat harga cabai cenderung lebih fluktuatif dan sensitif terhadap perubahan musiman atau cuaca. Secara keseluruhan, model GRU menunjukkan kinerja yang cukup baik dalam mempelajari pola harga dan menghasilkan prediksi yang mendekati nilai aktual, terutama untuk komoditas yang pola harganya lebih stabil.

### 3.3 Pemilihan Model Terbaik

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

columns = ['Bawang Merah', 'Bawang Putih', 'Cabai Merah', 'Cabai Rawit']

# List untuk simpan hasil
hasil_akhir = []

for i, col in enumerate(columns):
    # VAR â†’ evaluasi di data test
    y_var_actual = df_test[col]
    y_var_pred = forecast_df_real[col]
    rmse_var = np.sqrt(mean_squared_error(y_var_actual, y_var_pred))
    mape_var = mean_absolute_percentage_error(y_var_actual, y_var_pred) * 100

    # GRU
    rmse_gru = np.sqrt(mean_squared_error(y_test_inv[:, i], y_pred_inv[:, i]))
    mape_gru = mean_absolute_percentage_error(y_test_inv[:, i], y_pred_inv[:, i]) * 100

    hasil_akhir.append({
        'Komoditas': col,
        'VAR_RMSE': rmse_var,
        'VAR_MAPE (%)': mape_var,
        'GRU_RMSE': rmse_gru,
        'GRU_MAPE (%)': mape_gru
    })
tabel_perbandingan = pd.DataFrame(hasil_akhir)
```

```python
from tabulate import tabulate

print("TABEL PERBANDINGAN AKURASI MODEL VAR vs GRU (Data Test)")
print(tabulate(
    tabel_perbandingan,
    headers='keys',
    tablefmt='fancy_grid',
    showindex=False
))
```

```python
tabel_perbandingan['Model Terbaik (MAPE)'] = tabel_perbandingan.apply(
    lambda row: 'GRU' if row['GRU_MAPE (%)'] < row['VAR_MAPE (%)'] else 'VAR',
    axis=1
)

print("Kesimpulan Model Terbaik per Komoditas (berdasarkan MAPE terkecil):")
for i, row in tabel_perbandingan.iterrows():
    best_model = row['Model Terbaik (MAPE)']
    best_mape = min(row['VAR_MAPE (%)'], row['GRU_MAPE (%)'])
    print(f"Model terbaik untuk {row['Komoditas']} adalah **{best_model}** (MAPE: {best_mape:.2f}%)")
```

Berdasarkan hasil perbandingan akurasi antara model Vector Autoregression (VAR) dan Gated Recurrent Unit (GRU) pada data pengujian, terlihat bahwa model GRU secara konsisten menghasilkan nilai RMSE dan MAPE yang lebih rendah dibandingkan VAR untuk semua komoditas yang diprediksi. Pada komoditas bawang merah, selisih kinerja sangat mencolok, di mana VAR menghasilkan RMSE sebesar 13.038 dan MAPE lebih dari 26%, sedangkan GRU hanya menghasilkan RMSE sekitar 849 dan MAPE sebesar 1,73%. Hal serupa terjadi pada cabai merah dan cabai rawit, di mana GRU secara signifikan lebih akurat, dengan MAPE masing-masing 6,70% dan 5,37%, jauh di bawah MAPE VAR yang mencapai 29,19% dan 18,54%.

Bawang putih menjadi satu-satunya komoditas di mana kinerja VAR dan GRU relatif dekat, namun GRU tetap unggul dengan RMSE dan MAPE yang sedikit lebih rendah. Hasil ini menunjukkan bahwa GRU lebih mampu menangkap pola musiman, fluktuasi, serta dinamika harga yang kompleks, dibandingkan VAR yang mengandalkan hubungan linier antar variabel. Oleh karena itu, GRU dinilai sebagai model yang lebih baik dan lebih andal dalam melakukan prediksi harga komoditas bawang dan cabai di Kabupaten Boyolali.

### 3.4 Kelebihan dan Keterbatasan Model GRU

Model Gated Recurrent Unit (GRU) memiliki sejumlah kelebihan yang menjadikannya unggul dalam pemodelan deret waktu, terutama dalam konteks prediksi harga komoditas yang bersifat fluktuatif. Salah satu kelebihan utama GRU adalah kemampuannya dalam menangkap pola nonlinier dan dependensi jangka panjang antar waktu, yang sulit diakomodasi oleh model statistik tradisional seperti VAR. GRU juga lebih efisien dibandingkan model LSTM karena arsitekturnya lebih sederhana, sehingga membutuhkan waktu pelatihan yang lebih singkat dan sumber daya komputasi yang lebih ringan tanpa mengorbankan akurasi. Dalam penelitian ini, GRU terbukti memberikan hasil prediksi yang lebih akurat dibandingkan VAR, sebagaimana ditunjukkan oleh nilai RMSE dan MAPE yang lebih rendah pada semua komoditas yang diuji.

Namun demikian, model GRU juga memiliki keterbatasan. Salah satu keterbatasan utamanya adalah sifatnya sebagai *black-box*, yang membuat interpretasi terhadap hubungan antar variabel dan mekanisme internal model menjadi sulit dijelaskan secara eksplisit. Hal ini berbeda dengan model VAR yang memiliki keunggulan dalam interpretabilitas melalui uji kausalitas dan analisis impulse response. Selain itu, GRU sangat bergantung pada kualitas data pelatihan dan parameter arsitektur model (seperti jumlah unit, window size, dan epoch), sehingga jika tidak disesuaikan dengan baik, model dapat mengalami overfitting atau underfitting. GRU juga memerlukan data dalam jumlah besar dan terstruktur dengan baik agar mampu belajar secara optimal, sehingga pada dataset yang kecil atau tidak stabil, kinerjanya dapat menurun. Oleh karena itu, pemilihan GRU sebagai model prediktif harus disesuaikan dengan karakteristik data dan tujuan analisis.
